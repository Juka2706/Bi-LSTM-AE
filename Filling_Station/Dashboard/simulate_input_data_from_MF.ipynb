```python
import requests
import numpy as np
import pandas as pd

import time
import TimeSeries.Preprocessing as pp
url = 'http://127.0.0.1:8000/pushdata'
```

```python
#Load data

data_test_normal = pd.read_csv("..\Daten/Normal_12_03_24.csv")
data_test_normal =data_test_normal.fillna(value=0)

#widerstand_Transport_12_03_24
#Rondel_Widerstnd_12_03_24

data_test_anormal1 = pd.read_csv("..\Daten/Zuführ_wiederstand_12_03_24.csv")
data_test_anormal1 =data_test_anormal1.fillna(value=0)
data_test_anormal2 = pd.read_csv("..\Daten/widerstand_Transport_12_03_24.csv")
data_test_anormal2 =data_test_anormal2.fillna(value=0)
data_test_anormal3 = pd.read_csv("..\Daten/Rondel_Widerstnd_12_03_24.csv")
data_test_anormal3 =data_test_anormal3.fillna(value=0)


data_test_anormal = pd.concat([data_test_anormal1, data_test_anormal2, data_test_anormal3])

print(data_test_anormal.shape)
print(data_test_normal.shape)
```

```python
downsampling_factor=64
input_window=64
```

# Stream

```python
#stream_data=data_test_anormal
stream_data=pd.concat([data_test_normal, data_test_anormal])
```

```python
offset=0
input_window=64
window_to_send = input_window*downsampling_factor
for i in range(len(stream_data)):
    offset+=window_to_send
    test_data = np.array(stream_data.iloc[0+offset:window_to_send+offset,:])
    x = requests.post(url, json = test_data.tolist())
    if x.text[0] != 'null':
        print(x.text)

    time.sleep((window_to_send*6)/5000)

    if offset >= len(stream_data)-window_to_send:
        offset=0
```
